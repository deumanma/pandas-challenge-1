## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

subtotal_column = clean_df[['unit_price','qty']]
subtotal_column['subtotal'] = clean_df['unit_price'] * clean_df['qty']
subtotal_column
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.


def calculate_shipping_price(row):
    if row > 50:
        return 7 * row
    else:
        return 10 * row

    
    
clean_df = clean_df[['unit_price', 'unit_weight', 'qty']].copy()

clean_df['total_weight'] = clean_df['unit_weight'] * clean_df['qty']



clean_df['shipping_price'] = clean_df['total_weight'].apply(calculate_shipping_price)


clean_df
    
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

sales_tax_rate = 0.0925

clean_df['subtotal'] = clean_df['unit_price'] * clean_df['qty']

# Calculate total price including shipping and sales tax
clean_df['total_price'] = (clean_df['subtotal'] + clean_df['shipping_price']) * (1 + sales_tax_rate)



# Display the resulting DataFrame
clean_df
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

clean_df['line_cost'] = clean_df['unit_price'] * clean_df['qty'] + clean_df['shipping_price']
clean_df
```

```python
# Create a column for the profit of
# each line using line cost and line price

clean_df['line_profit'] = clean_df['total_price'] - clean_df['line_cost']
clean_df.head()
```
